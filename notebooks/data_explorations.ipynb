```python
from Bio import SeqIO
import os
```

```python
fasta_path = os.path.join(".." ,"downloads", "PF00257.fasta")
records = list(SeqIO.parse(fasta_path, "fasta"))
print(f"Total sequences: {len(records)}")
```

```python
for rec in records[:3]:
    print(rec.id)
    print(rec.description)
    print(str(rec.seq[:50]) + "…")
    print()
```

```python
lengths = [len(rec.seq) for rec in records]
print(f"Min length: {min(lengths)}")
print(f"Max length: {max(lengths)}")
print(f"Mean length: {sum(lengths)/len(lengths):.1f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(lengths, bins=30)
plt.title("PF00257 sequence‐length distribution")
plt.xlabel("Length (aa)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
